```python
path = "/home/thorpham/AI-Challenge/videos/cam_01.mp4"
track_file = "/home/thorpham/AI-Challenge/counting/track_results/cam_01.txt"
```

```python
tracks = {}
with open(track_file, 'r') as ft:
    lines = [line.strip('\n').split(' ') for line in ft]
    for line in lines:
        frameid = int(line[1])
        trackid = int(line[2])
        x1 = max(int(float(line[3])), 0)
        y1 = max(int(float(line[4])), 0)
        x2 = min(int(float(line[5])), 1280)
        y2 = min(int(float(line[6])), 720)
        cx = int((x1 + x2) / 2)
        cy = int((y1 + y2) / 2)


        label = line[7]
        if trackid in tracks:
            tracks[trackid]['endframe'] = frameid
            tracks[trackid]['bbox'].append([frameid, x1, y1, x2, y2, label])
            tracks[trackid]['tracklet'].append([cx, cy])
        else:
            tracks[trackid] = {'startframe': frameid,
                               'endframe': frameid,
                               'bbox': [[frameid, x1, y1, x2, y2, label]],
                               'tracklet': [[cx, cy]],
                               'video_idx': line[0]}
```

```python
cam = cv2.VideoCapture(path)
fr_id = 0
obj = tracks[5]
while True :
    fr_id += 1
    ret,frame = cam.read()
    bbox = obj["bbox"]
    for i in range(len(bbox)):
        idx = bbox[i][0]
        if idx== fr_id :
            xmin,ymin,xmax,ymax,clss =  bbox[i][1:]
            cv2.rectangle(frame,(xmin,ymin),(xmax,ymax),(0,0,255),2)
        cv2.imshow("im",frame)
        if cv2.waitKey(0) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break
cam.release()
```
